```python
#Summation without recursion
def summ(L):
  total=0
  for i in L:
    total +=i
  return total

def summR(L):
  if len(L) ==1:
    return L[0]
  else:
    return L[-1]+ summR(L[:-1])
```

```python
def loan_payment(principal, annual_rate, years):
    if principal <= 0 or annual_rate <= 0 or years <= 0:
        return "Error: All inputs must be positive"

    r = annual_rate / 12
    n = years * 12

    payment = (r * principal) / (1 - (1 + r) ** (-n))
    return payment

P = float(input("Enter loan amount: "))
annual_rate = float(input("Enter annual interest rate as decimal: "))
years = int(input("Enter number of years: "))

print(f"Monthly payment: ${loan_payment(P, annual_rate, years):.2f}")
```

```python
def getValues(n):
    values = []

    for i in range(n):
        while True:
            try:
                value = float(input("Enter value: "))
                if value < 0:
                    print("Invalid input. Value cannot be negative.")
                else:
                    values.append(value)
                    break
            except:
                print("Invalid input. Please enter a number.")

    return values

def calcAverage(alist):
    new_list = alist[3:]
    return sum(new_list) / len(new_list)

n = int(input("How many values? "))

values = getValues(n)

if n > 3:
    print("Average excluding first three values:", calcAverage(values))
else:
    print("Need more than 3 values.")
```

```python
class Greeting:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    def display(self, formal):
        if formal:
            print(self.last + ", " + self.first)
        else:
            print(self.first)

    def __str__(self):
        return self.first + " " + self.last

person = Greeting("Blanca", "Santos")

person.display(True)
person.display(False)
print(person)
```

```python
class Student:
    def __init__(self, name, number, courses):
        self.name = name
        self.number = number
        self.courses = courses

    def get_name(self):
        return self.name

    def get_courses(self):
        return self.courses

    def set_name(self, new_name):
        self.name = new_name

    def __len__(self):
        return len(self.courses)

    def enroll(self, course):
        if course not in self.courses:
            self.courses.append(course)

student1 = Student("Blanca", 12345, ["Finance"])

student1.enroll("Python")
student1.enroll("Math")

print(student1.get_name())
print(student1.get_courses())
print(len(student1))
```

```python
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

class Rectangle:
    def __init__(self, p1, p2, p3, p4):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.p4 = p4
        self.width = self.calculate_width()
        self.height = self.calculate_height()

    def calculate_width(self):
        return self.p1.distance(self.p2)

    def calculate_height(self):
        return self.p2.distance(self.p3)

    def area(self):
        return self.width * self.height

p1 = Point(0, 0)
p2 = Point(4, 0)
p3 = Point(4, 3)
p4 = Point(0, 3)

rect = Rectangle(p1, p2, p3, p4)

print("Width:", rect.width)
print("Height:", rect.height)
print("Area:", rect.area())
```

```python
class Employee:
    def __init__(self, name, wage, hours):
        self.name = name
        self.__wage = wage
        self.hours = hours

    def get_wage(self):
        return self.__wage

    def set_wage(self, wage):
        if wage >= 0:
            self.__wage = wage
        else:
            print("Error: wage cannot be negative")

    def calculate_pay(self):
        return self.__wage * self.hours

    def display_info(self):
        print("Name:", self.name)
        print("Hourly wage:", self.__wage)
        print("Hours worked:", self.hours)
        print("Weekly pay:", self.calculate_pay())

class Manager(Employee):
    def __init__(self, name, wage, hours, bonus):
        super().__init__(name, wage, hours)
        self.bonus = bonus

    def calculate_pay(self):
        return self.get_wage() * self.hours + self.bonus

    def display_info(self):
        super().display_info()
        print("Bonus:", self.bonus)

employee1 = Employee("Ana", 20, 40)
manager1 = Manager("Luis", 30, 40, 500)

employee1.display_info()
print()
manager1.display_info()
```
