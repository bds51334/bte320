```python
number = float(input("Enter a number:"))
the_other_number = float(input("Enter another number:"))
operator= input("Enter an operator (+, -, *, /):")
if operator == "+":
    print(number + the_other_number)
elif operator == "-":
    print(number - the_other_number)
elif operator == "*":
    print(number * the_other_number)
elif operator == "/":
    print(number / the_other_number)
else:
    print("Invalid operator")
```
